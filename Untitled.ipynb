```python
import os
import time
import json
import geocoder
import pandas as pd
from pprint import pprint
```

```python
last_ncov_json = sorted([e for e in os.listdir('.') if e.startswith('ncov_')], reverse=True)[0]
last_ncov_json
```

```python
country_df = pd.read_json(last_ncov_json)
country_df
```

```python
provinces_df = dict()
for idx in country_df.index:
    se = country_df.iloc[idx]
    if se['cities']:
        provinces_df[se['provinceName']] = pd.read_json(json.dumps(se['cities']))
pprint(provinces_df)
```

```python
for province in provinces_df:
    print(province)
    for city in provinces_df[province]["cityName"]:
        print(province, city)
        g = geocoder.baidu(u"{}{}".format(province, city), key='62c28637e66010289bdef04310267902')
        print(province, city, g.latlng)
```
